```python
!pip install -q openpyxl imbalanced-learn

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import (classification_report, confusion_matrix,
                             ConfusionMatrixDisplay, accuracy_score,
                             precision_score, recall_score, f1_score)
from sklearn.utils import resample
import warnings
warnings.filterwarnings('ignore')

print("Library berhasil diimport!")
```

```python
from google.colab import files
uploaded = files.upload()  # Upload file .xlsx dari lokal

df = pd.read_excel('health_dataset_200k_real_countries.xlsx')

print("=" * 55)
print("📌 STEP 1 | Informasi Dataset")
print("=" * 55)
print(f"Jumlah baris   : {df.shape[0]}")
print(f"Jumlah kolom   : {df.shape[1]}")
print("\nKolom & Tipe Data:")
print(df.dtypes)
print("\nPreview 5 Baris Pertama:")
df.head()
```

```python
print("=" * 55)
print("📌 STEP 2 | Deskripsi Fitur")
print("=" * 55)

# Ambil nama kolom dari dataset
kolom = df.columns

# Klasifikasi tipe fitur otomatis
jenis = []
keterangan = []

for col in kolom:
    if df[col].dtype == 'object':
        jenis.append("Nominal")
        keterangan.append(f"Fitur kategori: {col}")
    elif df[col].dtype in ['int64', 'float64']:
        jenis.append("Kontinyu")
        keterangan.append(f"Fitur numerik: {col}")
    else:
        jenis.append("Lainnya")
        keterangan.append(f"Fitur lain: {col}")

# Tandai target (misalnya 'label')
target_col = 'label'   # ⚠️ pastikan ini sesuai dengan kolom target kamu

for i, col in enumerate(kolom):
    if col == target_col:
        jenis[i] = "TARGET (Klasifikasi)"
        keterangan[i] = f"Label klasifikasi: {col}"

# Buat tabel deskripsi fitur
fitur_info = pd.DataFrame({
    "Nama Fitur": kolom,
    "Jenis": jenis,
    "Keterangan": keterangan
})

print(fitur_info.to_string(index=False))

print("\n📊 Statistik Deskriptif:")
print(df.describe())
```

```python
print("=" * 55)
print("📌 STEP 3 | Distribusi Kelas (Label/Output)")
print("=" * 55)

import matplotlib.pyplot as plt

# Pastikan kolom label sudah ada
target_col = 'Heart_Risk' # Mengubah 'label' menjadi 'Heart_Risk'

# Hitung distribusi kelas
class_dist = df[target_col].value_counts()

print(class_dist)
print(f"\nTotal data   : {len(df)}")
print(f"Jumlah kelas : {df[target_col].nunique()}")

print("\nProporsi Kelas:")
print((class_dist / len(df) * 100).round(2).astype(str) + '%')


# ==============================
# Visualisasi (AMAN tanpa error warna)
# ==============================
fig, ax = plt.subplots(figsize=(7, 4))

bars = ax.bar(class_dist.index.astype(str), class_dist.values)

# Tambahkan angka di atas bar
for bar, val in zip(bars, class_dist.values):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
            str(val), ha='center', va='bottom', fontweight='bold')

ax.set_title('Distribusi Kelas Label (Sebelum Balancing)', fontweight='bold')
ax.set_ylabel('Jumlah Data')
ax.set_xlabel('Kelas')
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()


# ==============================
# Cek imbalance otomatis
# ==============================
max_count = class_dist.max()
min_count = class_dist.min()

if min_count / max_count < 0.7:
    print("\n⚠️ Data TIDAK SEIMBANG → Disarankan oversampling.")
else:
    print("\n✅ Data sudah cukup seimbang.")
```

```python
print("=" * 55)
print("📌 STEP 4 | Pengecekan & Penanganan Missing Values")
print("=" * 55)

# ==============================
# Cek Missing Value
# ==============================
print("Missing Values per Kolom:")
missing = df.isnull().sum()
print(missing)

# ==============================
# Drop kolom identifier (otomatis)
# ==============================
# Deteksi kolom ID / nama (jika ada)
drop_cols = [col for col in df.columns if 'id' in col.lower() or 'name' in col.lower()]

df_model = df.drop(columns=drop_cols)

if len(drop_cols) > 0:
    print(f"\n✅ Kolom identifier dihapus: {drop_cols}")
else:
    print("\n✅ Tidak ada kolom identifier yang perlu dihapus")

# ==============================
# Pisahkan tipe data
# ==============================
num_cols = df_model.select_dtypes(include=['int64','float64']).columns
cat_cols = df_model.select_dtypes(include=['object']).columns

# Pastikan label tidak ikut imputasi kategori
target_col = 'label'
cat_cols = [c for c in cat_cols if c != target_col]

# ==============================
# Imputasi Missing Value
# ==============================
# Numerik → median
df_model[num_cols] = df_model[num_cols].fillna(df_model[num_cols].median())

# Kategorik → modus
for col in cat_cols:
    df_model[col] = df_model[col].fillna(df_model[col].mode()[0])

print("\n✅ Imputasi selesai (median untuk numerik, modus untuk kategorik)")

# ==============================
# Cek ulang missing value
# ==============================
print(f"\nMissing setelah imputasi: {df_model.isnull().sum().sum()}")
```

```python
print("=" * 55)
print("📌 STEP 5 | Klasifikasi Tipe Fitur")
print("=" * 55)

# Tentukan target
target_col = 'label'

fitur = []
tipe = []
keterangan = []

for col in df_model.columns:

    # TARGET
    if col == target_col:
        fitur.append(col)
        tipe.append("TARGET (Nominal)")
        keterangan.append("Label klasifikasi")

    # NOMINAL (kategori)
    elif df_model[col].dtype == 'object':
        fitur.append(col)
        tipe.append("Nominal")
        keterangan.append("Data kategori tanpa urutan")

    # NUMERIK
    elif df_model[col].dtype in ['int64','float64']:

        unique_val = df_model[col].nunique()

        # ORDINAL (nilai sedikit & berurutan)
        if unique_val <= 10:
            fitur.append(col)
            tipe.append("Ordinal")
            keterangan.append("Nilai terbatas dan berurutan")

        # KONTINYU
        else:
            fitur.append(col)
            tipe.append("Kontinyu")
            keterangan.append("Nilai numerik berkelanjutan")

    else:
        fitur.append(col)
        tipe.append("Lainnya")
        keterangan.append("Tipe data lain")

# Buat tabel
tabel_fitur = pd.DataFrame({
    "Fitur": fitur,
    "Tipe": tipe,
    "Keterangan": keterangan
})

print(tabel_fitur.to_string(index=False))
```

```python
print("=" * 55)
print("📌 STEP 6 | Encoding Fitur Kategorik")
print("=" * 55)

from sklearn.preprocessing import LabelEncoder

# Tentukan target
target_col = 'Heart_Risk'

# ==============================
# Encoding fitur kategorik (otomatis)
# ==============================
le_dict = {}

cat_cols = df_model.select_dtypes(include=['object']).columns

for col in cat_cols:
    if col != target_col:
        le = LabelEncoder()
        df_model[col] = le.fit_transform(df_model[col])
        le_dict[col] = dict(zip(le.classes_, le.transform(le.classes_)))

        print(f"\nEncoding '{col}':")
        print(le_dict[col])

# ==============================
# Encoding target (jika masih object)
# ==============================
if df_model[target_col].dtype == 'object':
    le_target = LabelEncoder()
    df_model[target_col] = le_target.fit_transform(df_model[target_col])

    print(f"\nEncoding TARGET '{target_col}':")
    print(dict(zip(le_target.classes_, le_target.transform(le_target.classes_))))
else:
    print(f"\nTARGET '{target_col}' sudah numerik (tidak perlu encoding)")

# ==============================
# Hasil akhir
# ==============================
print("\nDataset setelah encoding (5 baris pertama):")
print(df_model.head())
```

```python
print("=" * 55)
print("📌 STEP 7 | Normalisasi Data (MinMaxScaler)")
print("=" * 55)

from sklearn.preprocessing import MinMaxScaler
import seaborn as sns
import matplotlib.pyplot as plt

# ==============================
# Tentukan target
# ==============================
target_col = 'Heart_Risk'

# ==============================
# Pisahkan fitur & target
# ==============================
feature_cols = [c for c in df_model.columns if c != target_col]

X_raw = df_model[feature_cols].copy()
y_raw = df_model[target_col].copy()

# ==============================
# Normalisasi
# ==============================
scaler = MinMaxScaler()
X_scaled = pd.DataFrame(scaler.fit_transform(X_raw), columns=feature_cols)

print("Hasil normalisasi (5 baris pertama):")
print(X_scaled.head().to_string())

print("\nMin setelah normalisasi:")
print(X_scaled.min().to_dict())

print("\nMax setelah normalisasi:")
print(X_scaled.max().to_dict())


# ==============================
# Pairplot (otomatis ambil 5 fitur numerik)
# ==============================
print("\n📊 Membuat Pairplot...")

# Ambil maksimal 5 fitur numerik biar tidak berat
num_cols = X_scaled.select_dtypes(include=['float64','int64']).columns[:5]

plot_df = X_scaled[num_cols].copy()
plot_df['Kelas'] = y_raw.values

pair = sns.pairplot(
    plot_df,
    hue='Kelas',
    diag_kind='kde',
    plot_kws={'alpha': 0.4, 's': 15}
)

pair.fig.suptitle('Pairplot Dataset Health (Fitur Terpilih)',
                  y=1.02, fontsize=13, fontweight='bold')

plt.show()


# ==============================
# Penjelasan otomatis
# ==============================
print("""
📝 Makna Pairplot:
• Diagonal (KDE): menunjukkan distribusi tiap fitur untuk setiap kelas.
• Jika distribusi antar kelas terpisah → fitur tersebut bagus untuk klasifikasi.
• Jika saling tumpang tindih → fitur kurang berpengaruh.

• Scatter plot (off-diagonal):
  menunjukkan hubungan antar fitur.
• Pola cluster yang terpisah → model KNN lebih mudah membedakan kelas.

Kesimpulan:
Semakin jelas pemisahan antar kelas pada plot,
semakin baik performa model klasifikasi.
""")
```

```python
print("=" * 55)
print("📌 STEP 8 | Penyeimbangan Kelas — Random Oversampling")
print("=" * 55)

from sklearn.utils import resample
import matplotlib.pyplot as plt

# ==============================
# Distribusi sebelum balancing
# ==============================
print("Distribusi sebelum balancing:")
print(y_raw.value_counts())

# Gabungkan fitur + target
df_full = X_scaled.copy()
df_full['target'] = y_raw.values

# ==============================
# Oversampling
# ==============================
classes = df_full['target'].value_counts()
max_count = classes.max()

print(f"\nJumlah target oversampling: {max_count} per kelas")

balanced_parts = []

for cls in classes.index:
    cls_df = df_full[df_full['target'] == cls]

    cls_resampled = resample(
        cls_df,
        replace=True,
        n_samples=max_count,
        random_state=42
    )

    balanced_parts.append(cls_resampled)

# Gabungkan & shuffle
df_balanced = (
    pd.concat(balanced_parts)
    .sample(frac=1, random_state=42)
    .reset_index(drop=True)
)

# Pisahkan kembali
X_bal = df_balanced.drop(columns=['target'])
y_bal = df_balanced['target']

# ==============================
# Distribusi setelah balancing
# ==============================
print("\nDistribusi setelah balancing:")
print(y_bal.value_counts())


# ==============================
# Visualisasi sebelum vs sesudah
# ==============================
fig, axes = plt.subplots(1, 2, figsize=(12, 4))

dist_before = y_raw.value_counts()
dist_after = y_bal.value_counts()

for ax, dist, title in zip(
    axes,
    [dist_before, dist_after],
    ['Sebelum Balancing', 'Setelah Oversampling']
):
    bars = ax.bar(dist.index.astype(str), dist.values)

    for bar, v in zip(bars, dist.values):
        ax.text(
            bar.get_x() + bar.get_width()/2,
            bar.get_height(),
            str(v),
            ha='center',
            va='bottom',
            fontweight='bold'
        )

    ax.set_title(title, fontweight='bold')
    ax.set_ylabel('Jumlah Data')
    ax.set_xlabel('Kelas')
    ax.grid(axis='y', alpha=0.3)

fig.suptitle('Distribusi Kelas: Sebelum vs Setelah Balancing',
             fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()
```

```python
print("=" * 55)
print("📌 STEP 9 | Split Dataset 80:20 (Stratified)")
print("=" * 55)

from sklearn.model_selection import train_test_split

# ==============================
# Split data (80:20)
# ==============================
X_train, X_test, y_train, y_test = train_test_split(
    X_bal, y_bal,
    test_size=0.2,
    random_state=42,
    stratify=y_bal
)

# ==============================
# Informasi jumlah data
# ==============================
print(f"Total data (setelah balancing) : {len(X_bal)}")
print(f"Data Training (80%)            : {len(X_train)}")
print(f"Data Testing  (20%)            : {len(X_test)}")

# ==============================
# Distribusi kelas
# ==============================
print("\nDistribusi Training:")
print(y_train.value_counts())

print("\nDistribusi Testing:")
print(y_test.value_counts())
```

```python
print("=" * 55)
print("📌 STEP 10 | Pencarian K Optimal — 5-Fold Cross Validation")
print("=" * 55)

from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import cross_val_score
import numpy as np
import matplotlib.pyplot as plt

# ==============================
# Range nilai K
# ==============================
k_range = range(1, 31)
cv_scores = []

# ==============================
# Cross Validation
# ==============================
for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)

    scores = cross_val_score(
        knn,
        X_train,
        y_train,
        cv=5,
        scoring='accuracy'
    )

    mean_score = scores.mean()
    cv_scores.append(mean_score)

    print(f"K={k:2d} | CV Accuracy = {mean_score:.4f}")

# ==============================
# K terbaik
# ==============================
best_k = k_range[np.argmax(cv_scores)]
best_cv_acc = max(cv_scores)

print(f"\n🏆 K Optimal = {best_k}")
print(f"📊 CV Accuracy terbaik = {best_cv_acc:.4f}")

# ==============================
# Plot hasil
# ==============================
plt.figure(figsize=(10, 4))

plt.plot(list(k_range), cv_scores, marker='o')
plt.axvline(best_k, linestyle='--')
plt.scatter(best_k, best_cv_acc)

plt.xlabel("Nilai K")
plt.ylabel("CV Accuracy (5-Fold)")
plt.title("Pencarian K Optimal pada KNN")

plt.grid()
plt.show()
```

```python
print("=" * 55)
print(f"📌 STEP 11 | Training KNN (K={best_k})")
print("=" * 55)

from sklearn.neighbors import KNeighborsClassifier

# ==============================
# Training model
# ==============================
knn_final = KNeighborsClassifier(n_neighbors=best_k)

knn_final.fit(X_train, y_train)

# ==============================
# Prediksi
# ==============================
y_pred = knn_final.predict(X_test)

# ==============================
# Informasi hasil
# ==============================
print(f"✅ Model KNN (K={best_k}) berhasil ditraining")
print(f"Jumlah data training : {X_train.shape[0]}")
print(f"Jumlah data testing  : {X_test.shape[0]}")
print(f"Jumlah fitur         : {X_train.shape[1]}")
```

```python
print("=" * 55)
print(f"📌 STEP 12 | Confusion Matrix (Testing Data)")
print("=" * 55)

from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# ==============================
# Ambil nama kelas otomatis
# ==============================
class_names = np.unique(y_test)

# ==============================
# Hitung confusion matrix
# ==============================
cm = confusion_matrix(y_test, y_pred)

# ==============================
# Visualisasi
# ==============================
fig, ax = plt.subplots(figsize=(7, 5))

disp = ConfusionMatrixDisplay(
    confusion_matrix=cm,
    display_labels=class_names
)

disp.plot(ax=ax, cmap='Blues', colorbar=True)

ax.set_title(f'Confusion Matrix — KNN (K={best_k})',
             fontsize=14, fontweight='bold')

plt.tight_layout()
plt.show()

# ==============================
# Tampilkan dalam bentuk tabel
# ==============================
print("\nConfusion Matrix (raw):")

cm_df = pd.DataFrame(
    cm,
    index=[f'Actual {c}' for c in class_names],
    columns=[f'Pred {c}' for c in class_names]
)

print(cm_df)

# ==============================
# Penjelasan otomatis
# ==============================
print("\n📝 Penjelasan Confusion Matrix:")
print("• Baris = kelas AKTUAL")
print("• Kolom = kelas PREDIKSI")
print("• Diagonal = prediksi BENAR")
print("• Di luar diagonal = kesalahan prediksi")

# Khusus untuk binary classification
if len(class_names) == 2:
    tn, fp, fn, tp = cm.ravel()

    print(f"""
Detail (Binary Classification):
• True Positive  (TP) = {tp}
• True Negative  (TN) = {tn}
• False Positive (FP) = {fp}
• False Negative (FN) = {fn}
""")
```

```python
print("=" * 55)
print(f"📌 STEP 13 | Evaluasi Model — Metrik Evaluasi")
print("=" * 55)

from sklearn.metrics import classification_report, accuracy_score, precision_score, recall_score, f1_score
import matplotlib.pyplot as plt
import numpy as np

# ==============================
# Ambil nama kelas otomatis
# ==============================
class_names = [str(c) for c in np.unique(y_test)]

# ==============================
# Classification Report
# ==============================
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=class_names))

# ==============================
# Hitung metrik
# ==============================
acc  = accuracy_score(y_test, y_pred)
prec = precision_score(y_test, y_pred, average='macro')
rec  = recall_score(y_test, y_pred, average='macro')
f1   = f1_score(y_test, y_pred, average='macro')

print(f"{'─'*40}")
print(f"  Accuracy  (Overall) : {acc:.4f}  ({acc*100:.2f}%)")
print(f"  Precision (Macro)   : {prec:.4f}")
print(f"  Recall    (Macro)   : {rec:.4f}")
print(f"  F1-Score  (Macro)   : {f1:.4f}")
print(f"{'─'*40}")

# ==============================
# Visualisasi metrik
# ==============================
metrics_labels = ['Accuracy', 'Precision\n(macro)', 'Recall\n(macro)', 'F1-Score\n(macro)']
metrics_values = [acc, prec, rec, f1]

fig, ax = plt.subplots(figsize=(8, 4))

bars = ax.bar(metrics_labels, metrics_values)

for bar, val in zip(bars, metrics_values):
    ax.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f'{val:.3f}',
        ha='center',
        va='bottom',
        fontweight='bold'
    )

ax.set_ylim(0, 1.1)
ax.set_ylabel('Score')
ax.set_title(f'Metrik Evaluasi Model KNN (K={best_k})')
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.show()

# ==============================
# Interpretasi otomatis
# ==============================
print(f"""
📝 Interpretasi Hasil:
• Accuracy  {acc:.2%} → Persentase prediksi benar dari seluruh data uji
• Precision {prec:.2%} → Ketepatan model saat memprediksi suatu kelas
• Recall    {rec:.2%} → Kemampuan model menemukan semua data suatu kelas
• F1-Score  {f1:.2%} → Keseimbangan antara precision dan recall

Kesimpulan:
Semakin tinggi nilai metrik (mendekati 1), semakin baik performa model.
""")
```
